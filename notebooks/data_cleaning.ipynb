```python

def calculateSkewness(df):
    dataTypeDict = dict(df.dtypes)
    for column in df.columns:
        # print(dataTypeDict[column])
        # print(df[column])
        if dataTypeDict[column] != 'object':
            newDf = df.copy()
            newDf[column] = newDf[column].fillna(newDf[column].median())
            return scipy.stats.skew(newDf[column], axis=0, bias=True)

calculateSkewness(df)
```
