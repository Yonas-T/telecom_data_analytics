```python
def fillMissingValues(df):
    for columns in df.columns:
        if df[columns].dtype == 'object':
            df[columns].fillna(value='Missed', inplace=True)
        else:
            df[columns].fillna(value=df[columns].mean(), inplace=True)

fillMissingValues(df)
```

```python
def aggregate_per_user(df):
    df_wo_bearerId = df.drop(['Bearer Id'], axis=1)
    df_bearerID = df[['Bearer Id', 'MSISDN/Number']]


    dups_color = df_bearerID.pivot_table(columns=['MSISDN/Number'], aggfunc='size')
    # print (dups_color)
    # print(pd.DataFrame(dups_color, columns = ['Bearer Id']))
    # print(pd.DataFrame(dups_color, columns = ['Bearer Id']).iloc[0])
    bearerCount = pd.DataFrame(dups_color, columns = ['Bearer Id'])
    bc = bearerCount.reset_index(drop=True)

    cc = df_wo_bearerId.to_numpy()
    unq, unq_inv = np.unique(cc[:, 0], return_inverse=True)
    out = np.zeros((len(unq), cc.shape[1]), dtype=cc.dtype)
    out[:, 0] = unq
    

    np.add.at(out[:, 1:], unq_inv, cc[:, 1:])
    
    # out  
    dfconverted = pd.DataFrame(out, columns = df_wo_bearerId.columns.tolist())

    horizontal_stack = pd.concat([bc, dfconverted], axis=1)
    print(horizontal_stack['Youtube DL (Bytes)'] + horizontal_stack['Youtube UL (Bytes)'])
    return horizontal_stack

aggregate_per_user(df)
```
