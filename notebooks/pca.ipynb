```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import sklearn
```

```python
cleaned_df = pd.read_csv('../data/cleaned_data_source.csv')
cleaned_df
```

```python
cleaned_numpy = cleaned_df.to_numpy()
print(cleaned_df.shape[0])

mean_list = []

for i in range(cleaned_df.shape[0]):
    np.mean(cleaned_numpy[i, :])
    mean_list.append(np.mean(cleaned_numpy[i, :]))
# mean_x = np.mean(cleaned_numpy[0,:])
# mean_y = np.mean(cleaned_numpy[1,:])
# mean_z = np.mean(cleaned_df[2,:])

mean_vector = np.array(mean_list)

print('Mean Vector:\n', mean_vector)
```

```python
scatter_matrix = np.zeros((cleaned_df.shape[0],cleaned_df.shape[1]))
for i in range(cleaned_numpy.shape[1]):
    scatter_matrix += (cleaned_numpy[:,i].reshape(cleaned_df.shape[0],1) - mean_vector).dot((cleaned_numpy[:,i].reshape(cleaned_df.shape[0],1) - mean_vector).T)
print('Scatter Matrix:\n', scatter_matrix)
```

```python
cov_mat = np.cov([cleaned_numpy[0,:],cleaned_numpy[1,:],cleaned_numpy[2,:]])
print('Covariance Matrix:\n', cov_mat)
```
